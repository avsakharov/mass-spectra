```python
from spectrum import Spectrum

# Пример данных
mz = [3.0, 3.5, 4.0, 4.5, 5.0]
intensities = [10, 20, 15, 5, 10]
substance_name = "Sample Substance"
metadata = {"id": "12345", "molecular_formula": "H2O", "molecular_weight": 18.01528}

# Создаем экземпляр класса Spectrum
spectrum = Spectrum(mz, intensities, substance_name, **metadata)
# Нормализация спектра
spectrum.normalize()
# Построение графика
spectrum.plot()
# Построение графика в виде столбцов
spectrum.barplot()
```
